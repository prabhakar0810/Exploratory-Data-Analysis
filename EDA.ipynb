```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample=pd.read_csv("SampleSuperstore.csv")
sample.head()
```

```python
sample.describe()
```

```python
sample.info()
```

```python
plt.figure(figsize=(14,5)).set_facecolor("lightblue")
sample["Sub-Category"].value_counts().sort_values(ascending=False).plot(kind="bar",color=["blue","green","yellow"])
plt.xticks(rotation=30,color="black",fontsize=16)
plt.yticks(color="blue")
plt.xlabel("Ship mode",color="green",fontsize=20)
plt.ylabel("Sales",color="green",fontsize=20)
plt.title("Shipmode and Sales",color="black",fontsize=25)
#plt.grid()
plt.legend("Sub-Category")
plt.show()
```

```python
pi=sample.groupby(["Sub-Category"])["Sales"].sum().sort_values(ascending=False).head(10)
pi
```

```python
sample["Sub-Category"].value_counts().sort_values(ascending=False).plot(kind="pie")
```

```python
#subplots
fig,ax=plt.subplots(2,2,figsize=(20,20))
ax[0,0].bar(sample["Sub-Category"],sample["Sales"],color="black")
ax[0,0].set_title("Sales vs sub category")
ax[0,0].tick_params(axis="x",rotation=30)

#plot_2
ax[0,1].bar(sample["Sub-Category"],sample["Profit"],color="black")
ax[0,1].set_title("Profit vs sub category")
ax[0,1].tick_params(axis="x",rotation=30)


#plot_3
ax[1,0].scatter(sample["Sub-Category"],sample["Sales"],color="black",marker="*",edgecolor="red")
ax[1,0].set_title("Sales vs sub category")
ax[1,0].tick_params(axis="x",rotation=30)


#plot_4
ax[1,1].bar(sample["Sub-Category"],sample["Profit"],color="black")
ax[1,1].set_title("Profit vs sub category")
ax[1,1].tick_params(axis="x",rotation=30)
```

```python
corr=sample[["Postal Code","Sales","Quantity","Discount","Profit"]].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap="magma",linewidths=2,linecolor="red")
```
